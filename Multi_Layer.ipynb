```python
#덧셈 결과
import numpy as np
from tensorflow import keras
from sklearn.model_selection import train_test_split
import tensorflow as tf

# 학습 데이터 생성
data_size = 1000
train_data = np.random.randint(100, size=(data_size, 2))
train_ans = (train_data[:, 0] + train_data[:, 1]).reshape(-1, 1)

# 모델 정의
model = keras.models.Sequential()
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(20, activation='relu'))
model.add(keras.layers.Dense(5, activation='relu'))
model.add(keras.layers.Dense(1, activation='elu'))

model.compile(loss='mse', optimizer=keras.optimizers.Adam(0.00029), metrics=['accuracy'])
model.fit(train_data, train_ans, batch_size=1, epochs=20)

z = np.array([10, 20]).reshape(1, 2)
q = model.predict(z, batch_size=1)
print(f"덧셈결과 {z[0][0]} + {z[0][1]} = {q[0][0]}")
```

```python
import numpy as np
from tensorflow import keras
from sklearn.model_selection import train_test_split
import tensorflow as tf

# 학습 데이터 생성
data_size = 5000
train_data = np.random.randint(1, 100, size=(data_size, 2))
train_ans = (train_data[:, 0] * train_data[:, 1]).reshape(-1, 1)

# 모델 정의
model = keras.models.Sequential()
model.add(keras.layers.Dense(10, activation='sigmoid'))
model.add(keras.layers.Dense(20, activation='sigmoid'))
model.add(keras.layers.Dense(5, activation='elu'))
model.add(keras.layers.Dense(1))

model.compile(loss='mse', optimizer=keras.optimizers.Adam(0.00029), metrics=['accuracy'])
model.fit(train_data, train_ans, batch_size=5, epochs=40)

z = np.array([20, 10]).reshape(1, 2)
q = model.predict(z, batch_size=1)
print(f"곱셈결과 {z[0][0]} * {z[0][1]} = {q[0][0]}")
```

```python
import numpy as np
from tensorflow import keras

# 학습 데이터 생성
data_size = 100000
train_data = np.random.randint(1, 100, size=(data_size, 2))
train_ans = np.log(train_data[:, 0] * train_data[:, 1]).reshape(-1, 1)

# 로그를 취한 데이터를 사용하여 모델 정의 및 학습
model = keras.models.Sequential()
model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(64, activation='elu'))
model.add(keras.layers.Dense(1, activation='linear'))

model.compile(loss='mse', optimizer=keras.optimizers.Adam(0.00029), metrics=['accuracy'])
model.fit(train_data, train_ans, batch_size=50, epochs=20)

# 예측
z = np.array([40, 10]).reshape(1, 2)
q_log = model.predict(z, batch_size=1)
q = np.exp(q_log)


print(f"곱셈결과 {z[0][0]} * {z[0][1]} = {q[0][0]}")
```

```python
import numpy as np
from tensorflow import keras
from sklearn.model_selection import train_test_split
import tensorflow as tf

# 학습 데이터 생성
data_size = 1000
train_data = np.random.randint(1,100, size=(data_size, 2))
train_ans = (train_data[:, 0] / train_data[:, 1]).reshape(-1, 1)

# 모델 정의
model = keras.models.Sequential()
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(20, activation='relu'))
model.add(keras.layers.Dense(5, activation='elu'))
model.add(keras.layers.Dense(1, activation='linear'))

model.compile(loss='mse', optimizer=keras.optimizers.Adam(0.00029), metrics=['accuracy'])
model.fit(train_data, train_ans, batch_size=1, epochs=30)

z = np.array([20, 10]).reshape(1, 2)
q = model.predict(z, batch_size=1)
print(f"나눗셈결과 {z[0][0]} / {z[0][1]} = {q[0][0]}")
```
